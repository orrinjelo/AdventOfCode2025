```python
def isRepeatable(i):
    s = str(i)
    for j in range(1, len(s)//2+1):
        seg = s[:j]
        # 212121 -> j,  seg,  o_max
        #           1   2     6
        #           2   21    3
        #           3   212   2
        if len(s) % j != 0:
            continue
        rep = True
        for o in range(len(s)//j):
            if seg != s[o*j:(o+1)*j]:
                # print(seg, s[o*j:(o+1)*j])
                rep = False
                break
        if rep == True:
            return True
    return False
```

```python
print(isRepeatable('11111'))
print(isRepeatable('121212'))
print(isRepeatable('123123123'))
print(isRepeatable('1212121'))
print(isRepeatable('12312312'))
print(isRepeatable('22222122'))
```

```python
def isTwiced(i):
    s = str(i)
    if len(s) % 2 != 0:
        return False
    seg = s[:len(s)//2]
    seg2 = s[len(s)//2:]
    if seg == seg2:
        return True
    return False
```

```python
print(isTwiced('11111'))
print(isTwiced('111111'))
print(isTwiced('121121'))
print(isTwiced('123123123'))
print(isTwiced('1212121'))
print(isTwiced('12312312'))
print(isTwiced('22222122'))
```

```python
def part1(inp):
    count = 0
    for entry in inp.split(','):
        [s, e] = entry.split('-')
        s = int(s)
        e = int(e)
        for i in range(s,e+1):
            if isTwiced(i):
                count += i
                # print(i)
    return count
```
