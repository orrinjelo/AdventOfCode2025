```python
def parse(inp):
    mmap = []
    for line in inp.split('\n'):
        l = []
        for c in line:
            l.append(9 if c == 'S' else 2 if c == '^' else 0)
        mmap.append(l)
    return mmap
```

```python
def part1(mm):
    global nsplit
    m = mm[:]
    y = 0
    x = m[0].index(9)
    nsplit = 0

    def prop(m,x,y):
        global nsplit
        if y == len(m):
            return
        if m[y][x] == 1:
            return
        if m[y][x] == 2:
            if x != 0:
                m[y][x-1] = 1
                nsplit += 1
                prop(m,x-1,y+1)
            if x != len(m[0])-1:
                m[y][x+1] = 1
                nsplit += 1
                prop(m,x+1,y+1)
        else:
            m[y][x] = 1
            prop(m,x,y+1)

    prop(m, x, y)
    return m, nsplit//2
```

```python
from copy import deepcopy

def part2(mm):
    global frames, f
    m = mm[:]
    y = 0
    x = m[0].index(9)
    frames = []
    f = 0
    def prop(mm,x,y):
        global frames, f
        m = deepcopy(mm)
        if y == len(m):
            f += 1
            frames.append(m)
            return
        # if m[y][x] == 1:
        #     return
        if m[y][x] == 2:
            if x != 0:
                m[y][x-1] = 1
                prop(m,x-1,y+1)
            if x != len(m[0])-1:
                m[y][x+1] = 1
                prop(m,x+1,y+1)
        else:
            m[y][x] = 1
            prop(m,x,y+1)

    prop(m, x, y)
    return frames, len(frames), f
```
