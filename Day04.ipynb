```python
def parse(inp):
    # Parse data
    data = []
    for line in inp.split():
        row = []
        for c in line.strip():
            row.append(1 if c == '@' else 0)
        data.append(row)
    ndata = np.array(data)
    return ndata

def pick(ndata):
    filtered = np.zeros_like(ndata)
    for r in range(len(ndata)):
        for c in range(len(ndata[r])):
            s = 0
            for x in range(-1,2):
                for y in range(-1,2):
                    if r+x >= 0 and r+x < len(ndata) and c+y >= 0 and c+y < len(ndata[r]):
                        if ndata[r+x][c+y] == 1 and not (x == 0 and y == 0) :
                            s += 1
            if ndata[r][c] == 1:
                filtered[r][c] = s
            else:
                filtered[r][c] = 9
    return filtered

# def step(ndata, filtered):
#     ndata[filtered < 4] = 0

def part2(inp):
    state = parse(inp)
    prev_state = np.ones_like(state)
    
    count = 0
    while not np.all(state == prev_state):
        filtered = pick(state)
        count += np.sum(filtered < 4)
        prev_state = np.copy(state)
        state[filtered < 4] = 0
    return count
```
